Choosing the Number of clusters

WCSS: Within Cluster sum of squares

```python
# finding wcss value for diffrent number of cluster
```

```python
wcss = []
```

```python
for i in range(1, 11):
    # Create and fit the KMeans model
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=42)
    kmeans.fit(X)
    # Append the inertia (within-cluster sum of squares) to the list
    wcss.append(kmeans.inertia_)
```

```python
print(f"Length of wcss: {len(wcss)}")
```

```python
if len(wcss) != 10:
    print("Adjusting wcss to have 10 elements...")
    # Adjust wcss here if necessary, this is just an example
    wcss = wcss[:10]
```

```python
sns.set()
plt.plot(range(1, len(wcss) + 1), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of Clustering Model

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
# return a label for each point based on their cluster
```

```python
Y = kmeans.fit_predict(X)
print(Y)
```

Visualizing the Clusters


```python
#plotting grapgh
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='green', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='green', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='green', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='green', label='Cluster 5')
```
